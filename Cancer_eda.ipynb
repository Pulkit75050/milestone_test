```python
sns.histplot(df['AGE'], bins = 10, kde = True, palette = 'salmon')

# here mostly the data has equal numbers of the age groups. But the 70-80 age group pouplation has the most of the samples.
# data is normaly skewed.
```

```python
plt.pie(x= df.LUNG_CANCER.value_counts(),autopct="%.3f", shadow =True,colors = ("pink","green"))
# half of the sample is suffering from Lung Cancer.
```

```python
pd.crosstab(df.ALCOHOL_CONSUMING,df.GENDER).plot(kind = "bar",stacked = True)
```

```python
print(df.groupby(["LUNG_CANCER","GENDER"])["SMOKING"].value_counts())
# here there are major 2 insights
# 1. females have more cancer as compared males.
# 2. most of females who DON'T smokes have more lung cancer but on the other side males who smoke more and have more lung cancer.
```

```python
sns.boxplot(x= "GENDER", y ="AGE",data=df,palette="viridis") 

# no outlier in the age of men and women.
```

```python
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot= True)

# no multicollinearity between the variables i.e. no independent variable is dependent of the other independent variables.
```

```python
sns.swarmplot(x="ALCOHOL_CONSUMING",y="AGE",hue="GENDER",data=df)
```

```python
# Q. find the no. of the samples who do both smoking and drinking but they have the lung cancer.

df[df["LUNG_CANCER"]=="YES"][df["SMOKING"]==1][df["ALCOHOL_CONSUMING"]==1]
# so there are 406 samples who have lung cancer and they smoke and drink.
```

```python
#  Q. find the number of women who has lung cancer and there age is above then the average age.

df[df["GENDER"]=="F"][df["LUNG_CANCER"]=="YES"][df["AGE"] > df["AGE"].mean()]

# hence there are 355 women who have cancer and they are above the avg age of the womrn.
```

```python
# Q. find out the percentage of the person who drink and smoke but don't have lung cancer.

df1 = df[df["LUNG_CANCER"]=="NO"][df["SMOKING"]==1][df["ALCOHOL_CONSUMING"]==1]

percentage = (df1.shape[0]*100)/df.shape[0]
percentage
```
